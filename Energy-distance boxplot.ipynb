```python

b_arr=numpy.zeros((21,21,21,2,2,2))
for i in data_set:
    for j in i:
        for k in j:
            tx = k.coords['x'].data
            ty = k.coords['y'].data
            tz = k.coords['z'].data
            b_arr[int(tx)+10,int(ty)+10,int(tz)+10,:,:,:] = math.sqrt(tx**2+ty**2+tz**2)
b_arr.data
```

```python
d_arr=numpy.reshape(b_arr,numpy.size(b_arr))
```

```python
def cond(ulim,d,b):
    return(d<=ulim) & (d_arr>ulim-1) & (b<0)
```
